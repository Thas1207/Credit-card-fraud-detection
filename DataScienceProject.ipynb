```python
train = pd.read_csv("fraudTrain.csv", parse_dates=['trans_date_trans_time',]) # change the dtype for date and time
test = pd.read_csv("fraudTest.csv", parse_dates=['trans_date_trans_time',]) # change the dtype for date and time

df1 = pd.concat([train,test],ignore_index=True)
df1.drop('Unnamed: 0',axis=1,inplace=True)
df1 = df1.drop(['street', 'zip','lat', 'long', 'city_pop', 'job','dob', 'trans_num','unix_time',
       'merch_lat', 'merch_long'], axis=1)
```

```python
df1.info()
```

```python
df1["hour"] = df1.trans_date_trans_time.dt.hour

df1['hourEnc'] = 0
df1.loc[df1.hour < 5,'hourEnc'] = 1
df1.loc[df1.hour > 21,'hourEnc'] = 1
```

```python
def lastDayTransactionCount(x):
    temp = pd.Series(x.index, index = x.trans_date_trans_time, name='count_day').sort_index()
    count_day = temp.rolling('1d').count() - 1
    count_day.index = temp.values
    x['count_day'] = count_day.reindex(x.index)
    return x
def lastWeekTransactionCount(x):
    temp = pd.Series(x.index, index = x.trans_date_trans_time, name='count_week').sort_index()
    count_week = temp.rolling('7d').count() - 1
    count_week.index = temp.values
    x['count_week'] = count_week.reindex(x.index)
    return x
def lastMonthTransactionCount(x):
    temp = pd.Series(x.index, index = x.trans_date_trans_time, name='count_month').sort_index()
    count_month = temp.rolling('30d').count() - 1
    count_month.index = temp.values
    x['count_month'] = count_month.reindex(x.index)
    return x
```

```python
df1 = df1.groupby('cc_num').apply(lastDayTransactionCount)
```

```python
df1 = df1.groupby('cc_num').apply(lastWeekTransactionCount)
```

```python
df1 = df1.groupby('cc_num').apply(lastMonthTransactionCount)
```

```python
def timeDifference(x):
    x['time_diff'] = x.trans_date_trans_time - x.trans_date_trans_time.shift()
    return x
```

```python
df1 = df1.groupby('cc_num').apply(timeDifference)
```

### Algorithm Selection

Algorithms:

```python
from sklearn.linear_model import LogisticRegression # Logistic Regression
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import LabelEncoder,StandardScaler
from warnings import filterwarnings
filterwarnings('ignore')
from sklearn.metrics import accuracy_score,precision_score,confusion_matrix #Performance_metrics
from sklearn.metrics import roc_curve,roc_auc_score, mean_squared_error
from sklearn.neighbors import KNeighborsClassifier  # KNN
from sklearn.ensemble import RandomForestClassifier # Random Forest
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.datasets import make_classification
```

```python
le = LabelEncoder()
df['gender_le']  = le.fit_transform(df['gender'])
df2 = df.select_dtypes(exclude = 'object')
df2 = df2.loc[:149999]
df2.head()
```

```python
#Test-Train Split
x = df2.drop('is_fraud',axis=1)
y = df2['is_fraud']
df_number = df.select_dtypes(exclude = 'object')
Xtrain,Xtest,Ytrain,Ytest = train_test_split(x,y,train_size = 0.7 , random_state = 10)
df_number = df_number.drop('is_fraud',axis=1)
# Scaling the data
ss = StandardScaler()
Xtrain[df_number.columns] = ss.fit_transform(Xtrain[df_number.columns])
Xtest[df_number.columns] = ss.fit_transform(Xtest[df_number.columns])
Xtest.head()
```

### Model Selection

```python
#K-Nearest Neighbors
knn1=KNeighborsClassifier()
params = {'n_neighbors':[2,3,4,5,7]}
knn_model = GridSearchCV(knn1, params, cv=5) #finding the best k value for the data
knn_model.fit(Xtrain,Ytrain)
knn_model.best_params_
```

```python
#Random Forest- estimator
X, y = make_classification(n_samples=1000,
                           n_features=10,
                           n_informative=3,
                           n_redundant=0,
                           n_repeated=0,
                           n_classes=2,
                           random_state=0,
                           shuffle=False)
rfc1 = RandomForestClassifier(n_jobs=-1,max_features= 'sqrt' ,n_estimators=50, oob_score = True) 

param_grid = { 
    'n_estimators': [300,500,700,900],
    'max_features': ['auto', 'sqrt', 'log2']
}
```

```python
CV_rfc = GridSearchCV(estimator=rfc1, param_grid=param_grid, cv= 5) #finding best parameters and estimitors for random forest
CV_rfc.fit(X, y)
```

```python
print (CV_rfc.best_params_)
```

```python
#Logistic Regression Model
lr = LogisticRegression()
#KNN model
knn = KNeighborsClassifier(n_neighbors=4)
#Random Forest model
rfc = RandomForestClassifier(n_estimators=700)
```

### Model Fitting

```python
model_1 = lr.fit(Xtrain,Ytrain)
model_2 = knn.fit(Xtrain,Ytrain)
model_3 = rfc.fit(Xtrain,Ytrain)
```

```python
Ypred = model_1.predict(Xtest)  #logistic Regression
Ypred1 = model_2.predict(Xtest) #KNN
Ypred2 = model_3.predict(Xtest) #Random Forest
```

```python
lr_cm  = confusion_matrix(Ytest,Ypred)
knn_cm = confusion_matrix(Ytest,Ypred1)
rf_cm  = confusion_matrix(Ytest,Ypred2)
```

```python
lr_acc = lr.score(Xtest,Ytest); rfc_acc= rfc.score(Xtest,Ytest); knn_acc= knn.score(Xtest,Ytest)
accuracy=[lr_acc,rfc_acc,knn_acc]
#Logistic Regression
TP = lr_cm[1,1] ; TN = lr_cm[0,0] ;FP = lr_cm[0,1]; FN = lr_cm[1,0]
pr_lr= TP/(TP+FP) ; re_lr= ss_lr=TP/(FN+TP); (TP/ float(TP+FN)) ; sp_lr= (TN/ float(TN+FP))
#KNN
TP = knn_cm[1,1]; TN = knn_cm[0,0]; FP = knn_cm[0,1]; FN = knn_cm[1,0]
pr_knn= TP/(TP+FP) ; re_knn=TP/(FN+TP);  ss_knn= (TP/ float(TP+FN)); sp_knn= (TN/ float(TN+FP))
#Random Forest
TP = rf_cm[1,1]; TN = rf_cm[0,0]; FP = rf_cm[0,1]; FN = rf_cm[1,0]
pr_rf= TP/(TP+FP) ; re_rf=TP/(FN+TP); ss_rf= (TP/ float(TP+FN)); sp_rf= (TN/ float(TN+FP))
#Performance_metrices
accuracy=[lr_acc,rfc_acc,knn_acc]
precision = [pr_lr,pr_knn,pr_rf]
recall = [re_lr,re_knn,re_rf]
sensitivity =[ss_lr,ss_knn,ss_rf]
specificity =[sp_lr,sp_knn,sp_rf]
```

```python
models=['Logistic Regression','K Nearest Neighbour','Random Forest']
result_table = pd.DataFrame({'Models':models,'Accuracy':accuracy,'Precision':precision,'Recall':recall,
                             'Sensitivity':sensitivity,'Specificity':specificity})
result_table
```
